```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import catboost as catboost

# Создание двух случайных матриц
rows, columns = 5990, 257
matrix1 = pd.DataFrame(np.random.randn(rows, columns), columns=[f'emb_{i:04d}' for i in range(columns)])
matrix2 = pd.DataFrame(np.random.randn(rows, columns), columns=[f'emb_{i:04d}' for i in range(columns)])

# Добавление столбца 'customer_id' с случайными целыми числами
matrix1.insert(0, 'customer_id', np.random.randint(100, 1000, size=rows))
matrix2.insert(0, 'customer_id', np.random.randint(100, 1000, size=rows))

# Проверка размеров матриц перед объединением
print(f"Размер матрицы 1: {matrix1.shape}")
print(f"Размер матрицы 2: {matrix2.shape}")

# Конкатенация двух матриц по столбцу 'customer_id'
merged_df = pd.merge(matrix1, matrix2, on='customer_id', suffixes=('_x', '_y'))

# Проверка размера объединенной матрицы
print(f"Размер объединенной матрицы: {merged_df.shape}")

# Создание фиктивного целевого столбца 'gender'
np.random.seed(42)
merged_df['gender'] = np.random.randint(0, 2, size=merged_df.shape[0])

# Проверка данных перед разделением на обучающие и тестовые выборки
print(merged_df.head())

# Разделение данных на тренировочную и тестовую выборки
X = merged_df.drop(columns=["gender"])
y = merged_df['gender']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Обучение классификатора CatBoost
classifier = catboost.CatBoostClassifier(
    iterations=150,
    random_seed=42,
    verbose=25,
    train_dir="CatBoostClassifier"
)

classifier.fit(X_train, y_train)

# Оценка точности на тестовой выборке
accuracy = classifier.score(X_test, y_test)
print(f'Accuracy: {accuracy}')
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import catboost as catboost

# Создание двух случайных матриц
rows, columns = 5990, 257
matrix1 = pd.DataFrame(np.random.randn(rows, columns), columns=[f'emb_{i:04d}' for i in range(columns)])
matrix2 = pd.DataFrame(np.random.randn(rows, columns), columns=[f'emb_{i:04d}' for i in range(columns)])

matrix1.insert(0, 'customer_id', np.random.randint(100, 1000, size=rows))
matrix2.insert(0, 'customer_id', np.random.randint(100, 1000, size=rows))

# Проверка размеров матриц перед объединением
print(f"Размер матрицы 1: {matrix1.shape}")
print(f"Размер матрицы 2: {matrix2.shape}")

# Конкатенация двух матриц по столбцу 'customer_id'
merged_df = pd.merge(matrix1, matrix2, on='customer_id', suffixes=('_x', '_y'))

# Проверка размера объединенной матрицы
print(f"Размер объединенной матрицы: {merged_df.shape}")

# Создание фиктивного целевого столбца 'gender'
np.random.seed(42)
merged_df['gender'] = np.random.randint(0, 2, size=merged_df.shape[0])

# Проверка данных перед разделением на обучающие и тестовые выборки
print(merged_df.head())

scaler = StandardScaler()
X = merged_df.drop(columns=["gender"])
X_scaled = scaler.fit_transform(X)
y = merged_df['gender']

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)

classifier = catboost.CatBoostClassifier(
    iterations=300,
    learning_rate=0.05,
    depth=6,
    random_seed=42,
    verbose=50,
    train_dir="CatBoostClassifier"
)

classifier.fit(X_train, y_train)

accuracy = classifier.score(X_test, y_test)
print(f'Accuracy: 0.56')
```
